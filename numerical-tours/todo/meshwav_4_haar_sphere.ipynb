```python
U = randn(4)U(: , 1) = 1[U, R] = qr(U)
```

Initialize the forward transform.

```python
fw = fnj = length(f)
```

Extract the low pass component and apply the matrix U

```python
fj = fw(1: nj)fj = reshape(fj, [nj/ 4 4])fj = fj*U
```

Store back the coefficients.

```python
fw(1: nj) = fj(: )nj = nj/ 4
```
