You can apply functions to each entry of a matrix

```python
E = A./ C; % divisionE = sin(A); % sinus is applied to each entryE = abs(A + 1i*C); % modulus of each entry
```

You can modify matrices and arrays in various way.

```python
b = sort(b); % sort valuesb = b .* (b >2); % set to zeros (threshold) the values below 2b(3) = []; % suppress the 3rd entry of a vectorB = [b; b]; % create a matrix of size 2x4c = B(: , 2); % to access 2nd column
```

Is is possible to access directly the last entry of a vector using thekeyword |end| in Matlab and |$| in Scilab.The equivalent Scilab code is      b($-2:$) = 1;

```python
b(end-2: end) = 1; % to access the last entries
```

The equivalent Scilab code is      b = b($:-1:1);

```python
b = b(end: -1: 1); % reverse a vector
```
